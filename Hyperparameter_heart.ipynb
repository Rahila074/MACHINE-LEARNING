```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)
print(accuracy)
print(classification_rep)
```

```python
knn1=KNeighborsClassifier()
from sklearn.model_selection import GridSearchCV
```

```python
param={'n_neighbors':[3,5,7,9],'weights':['uniform','distance']}
clf=GridSearchCV(knn1,param,cv=10,scoring='accuracy')  #CV-->cross validation
clf.fit(x_train,y_train)
```

```python
print(clf.best_params_)
```

```python
knn2=KNeighborsClassifier(n_neighbors=9,weights='distance')
knn2.fit(x_train,y_train)
y_pred1=knn2.predict(x_test)
y_pred1
```

```python
print(accuracy_score(y_pred1,y_test)) 
```
