```python
x = df
x
```

```python
from sklearn.cluster import KMeans
lst=[]
for i in range(1,11):
    data=KMeans(n_clusters=i,init="k-means++",random_state=42)
    data.fit(x)
    lst.append(data.inertia_)
```

```python
plt.plot(range(1,11),lst)
plt.xlabel("number of clusters")
plt.ylabel("variance")
plt.show()
```
