```python
df1=pd.get_dummies(df[['Country']],drop_first=True)
df1
```

```python
df['Status']=df['Status'].replace({'Developing':0,'Developed':1})
df
```

```python
df2=pd.concat([df,df1],axis=1)
df2
```

```python
df2=df2.drop(['Country'],axis=1)
df2
```

```python
list=['Population','Hepatitis B','Total expenditure','GDP','BMI','Alcohol','Schooling','Polio','Diphtheria']
```

```python
for i in list:
  df2[i]=df2[i].fillna(df2[i].median())
```

```python
print(df2.isna().sum().to_string())
```

```python
df2.dtypes
```

```python
x=df2.drop(['Life expectancy'],axis=1)
y=df2['Life expectancy']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
model=LinearRegression()
dec=DecisionTreeRegressor()
ran=RandomForestRegressor()
data=[model,dec,ran]
```

```python
from sklearn.metrics import r2_score,mean_absolute_percentage_error
for i in data:
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print(i)
  print("r2_score is",r2_score(y_test,y_pred))
  print("MAPE is",mean_absolute_percentage_error(y_test,y_pred))
  print("***************************************************************************************************************")
```
