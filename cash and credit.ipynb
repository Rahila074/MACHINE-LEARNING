```python
df1=pd.get_dummies(df[["City","Exp Type"]],drop_first=True,dtype=int)
```

```python
dfe=pd.concat([df,df1],axis=1)
dfe
```

```python
dfe.drop(["City","Exp Type"],axis=1,inplace=True)
dfe
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder=OrdinalEncoder(categories=[['Silver', 'Gold', 'Platinum', 'Signature']])
dfe["Card Type"]=encoder.fit_transform(dfe[["Card Type"]])
```

```python
x = dfe.drop(["Amount"],axis=1)
y = dfe['Amount']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
        
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(y_test,y_pred)
acc
```
